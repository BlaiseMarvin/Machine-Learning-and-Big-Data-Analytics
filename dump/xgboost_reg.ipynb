```python
import xgboost as xgb
import pandas as pd
import numpy as np
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
x_train2 = pd.read_csv('optimum_subtrainX.csv')
y_train2 = pd.read_csv('optimum_subtrainY.csv')
x_test2 = pd.read_csv('optimum_subtestX.csv')
y_test2 = pd.read_csv('optimum_subtestY.csv')
```

```python
x_train2.columns
```

```python
x_train2 = x_train2.set_index('Unnamed: 0')
x_test2=x_test2.set_index('Unnamed: 0')
y_train2 = y_train2.set_index('Unnamed: 0')
y_test2 = y_test2.set_index('Unnamed: 0')
```

```python
reg = xgb.XGBRegressor(objective='reg:squarederror',seed=42,eval_metric='rmse',early_stopping_rounds=50)
```

```python
parameters2 = {
    'learning_rate':[0.001,0.01,0.05,0.1],
    'max_depth': [1,2,4,6,8,10],
    'reg_lambda':[1.0,10.0,15.0,20.0],
    'subsample': [0.4, 0.9,1.0],
    'colsample_bytree':[0.4, 0.9,1.0]
}
```

```python
gridsearch = GridSearchCV(
    estimator=xgb.XGBRegressor(objective='reg:squarederror',seed=42,eval_metric='rmse'),
    param_grid=parameters2,
    scoring='neg_root_mean_squared_error',
    verbose=5,
    cv=3,
    n_jobs=8
)
```

```python
gridsearch.fit(x_train2,y_train2)
```

```python
gridsearch.best_score_
```

```python
gridsearch.best_params_
```

```python
reg = xgb.XGBRegressor(objective='reg:squarederror',seed=42,eval_metric='rmse',early_stopping_rounds=50,
                      learning_rate=0.05,max_depth=2,reg_lambda=15.0)
```

```python
reg.fit(x_train2,y_train2,eval_set=[(x_test2,y_test2)])
```

```python
test_set = pd.read_csv('data/CS98XRegressionTest.csv')
```

```python
predictors = list(x_train2.columns)
```

```python
ts2 = test_set.copy()
```

```python
ts2 = ts2[predictors]
```

```python
test_set['pop'] = reg.predict(ts2)
```

```python
test_set = test_set.reindex(columns=['Id','pop'])
```

```python
test_set.to_csv('modelling4x1.csv', index=False)
```
