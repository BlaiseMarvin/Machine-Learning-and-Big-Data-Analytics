```python
# function that unpivots columns to rows while leaving other columns intact pandas dataframe
def unpivot(t):
    return t.melt(id_vars=[c for c in t.columns if c not in ['A', 'B']], value_vars=['A', 'B'], var_name='variable', value_name='value')
```

```python
def add_monotonic_id(df, id_col='id'):
    df[id_col] = range(1, len(df) + 1)
    return df
```
