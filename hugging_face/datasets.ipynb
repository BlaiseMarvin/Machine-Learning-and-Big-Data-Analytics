**Split long examples:**
- when examples are too long, you may want to split them into several smaller chunks. begin by creating a function that does that

```python
def chunk_examples(examples):
    chunks = []
    for sentence in examples['sentence1']:
        chunks += [sentence[i:i+50] for i in range(0,len(sentence),50)]
    return {"chunks":chunks}
```

```python
chunked_dataset = dataset2.map(\
    chunk_examples,batched=True,remove_columns=dataset2.column_names\
)
```

**data augmentation:**
- the map function could also be used for data augmentation. the following example generates additional words for a masked token in a sentence.....

the map function can be used for data augmentation. we're gonna test this using the roberta model in transformers' fillmaskpipeline:

```python
from random import randint
from transformers import pipeline
```

```python
fillmask = pipeline("fill-mask", model="roberta-base")
```

```python
mask_token = fillmask.tokenizer.mask_token
smaller_dataset = dataset2.filter(lambda e,i : i<100, with_indices=True)
```

Create a function to randomly select a word to mask in the sentence. The function should also return the original sentence and the two top replacements generated by roberta.

```python
def augment_data(examples):
    outputs = []
    for sentence in examples['sentence1']:
        words = sentence.split(' ')
        K = randint(1, len(words)-1)
        masked_sentence = " ".join(words[:K]  + [mask_token] + words[K+1:])
        predictions = fillmask(masked_sentence)
        augmented_sequences = [predictions[i]["sequence"] for i in range(3)]
        outputs += [sentence] + augmented_sequences
    return {"data": outputs}
```

```python
augmented_dataset = smaller_dataset.map(augment_data, batched=True, remove_columns=smaller_dataset.column_names, batch_size=8)
augmented_dataset[:9]["data"]
```

**Stream Data:**
- stream a dataset by loading it as an iterabledataset. this allows you to progressively iterate over a remote dataset without downloading it on disk and or over local data files. 

**Parallel reading with and without shards:**
- num_workers parallelizes reading. Sharding parallelizes where you read from. 
- case 1: no sharding and num_workers > 0
   - one or a few large arrow files, and all workers reading from the same file
   - what happens: workers run in parallel, but they seek into the same file, compete for the same disk bandwidth, and share the same file handles/inode
   - works, but doesn't scale perfectly.
   - bottleneck: disk contention | single-file i/o limits | os read-ahead is less effective
   - multiple people reading different chapters of the same giant book at the same time

- case 2: sharded dataset + num_workers
   - dataset split into many files (shards)
   - each shard contains independent rows
   - each worker reads different shards
   - fewer seeks
   - more sequential reads
   - better disk parallelism

- single file, many workers:
  - disk head thrashes
  - throughput plateaus early

- many shards, many workers

using datasets with apache spark:
- ds = Dataset.from_spark(spark_dataframe)
  - the spark workers write the dataset on disk in a cache directory as arrow files, and the dataset is loaded from there

- Alternatively, you can skip materializing by using IterableDataset.from_spark - which returns an iterable dataset
  - iterable dataset streams rows from spark without materializing them
  - What each actually does:
    - Dataset.from_spark(spark_df) => materialized, disk-backed dataset. 
    - Spark executes the dataframe, data is collected partition-by-partition
    - written to arrow shards on disk
    - exposed as normal hf dataset
    - with dataset you get Spark -> Parquet -> Arrow -> mmap

  - IterableDataset.from_spark => streaming, non-materialized dataset
    - spark executes the dataframe
    - rows are yielded as an iterator
    - nothing written to disk
    - data flows directly to training
  - You get:
     - zero disk usage
     - immediate start
     - can handle datasets larger than disk
 
- Caching:
  - when using Dataset.from_spark() => the resulting Dataset is cached; if you call Dataset.from_spark multiple times on the same dataframe it won't rerun the spark job that writes the datasets as arrow files on disk
  - can set the cache location by passing cache_dir =  to Dataset.from_spark().
