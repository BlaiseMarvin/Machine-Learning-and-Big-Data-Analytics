radial basis function is often used to quantify the similarity between points...core idea behind rbf is u have a fixed ref point and variable point...
- most commonly used rbf is the Gaussian rbf, whose output value decays exponentially as the input value moves away from the firxed point...
- for similarity matrices we're looking at..formula will be: “exp(–γ(xi – xj)²)” -> remember though that (xi – xj)² was already initially computed in the matrix as the mxm similarity matrix of euclidean distances
- Hyperparameter gamma determines how quickly the similarity measure decays as x moves away from the fixed point...in otherwords as the distance increases

from the eqn..: 
- a higher gamma implies the similarity deteriorates fast and hence a large distance results in a low similarity score with a high gamma. from the eqn gamma seems to be the inverse of standard deviation so a high gamma...narrow gaussian and hence faster deterioration

- High gamma → small σ² → narrow Gaussian → fast decay → only very close points get high similarity
- Low gamma → large σ² → wide Gaussian → slow decay → distant points still get meaningful similarity
- exp( -||xi - xj||² / 2σ² ) | the full gaussian/normal distribution kernel
- rbf kernel implemented by sklearn: exp( -γ · ||xi - xj||² )
- γ = 1 / 2σ²
