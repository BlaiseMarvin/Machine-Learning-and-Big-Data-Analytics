```python
G = nx.Graph()
G.graph["Name"] = "My Graph"

G.add_nodes_from([
    ("A", {"Age":19, "Gender":"F"}),
    ("B", {"Age":18, "Gender":"M"}),
    ("C", {"Age":22, "Gender":"M"}),
    ("D", {"Age":21, "Gender":"M"}),
    ("E", {"Age":20, "Gender":"F"})
])

G.add_edges_from([
    ("A","C",{"weight":1}),
    ("B","C",{"weight":0.5}),
    ("B","D",{"weight":0.6}),
    ("C","D",{"weight":0.8}),
    ("D","E",{"weight":1})
])

pos = {
    "A" : (1, 5),
    "B": (4.5, 6.6),
    "C": (3.6, 1.4),
    "D": (5.8, 3.5),
    "E": (7.9, 3.6)
} # positions are defined as x and y coordinates

# define the positions of the node labels
# - handtyping the positions of the node labels
pos_node_attributes = {
    "A": (-0.2, 5),
    "B": (4.5, 7.5),
    "C": (2.4, 1.4),
    "D": (5.8, 2.5),
    "E": (9.1, 3.6)
} # these are the handtyped actual positions (x.y) coordinates of the node labels we want to add to the graph

# draw graph
nx.draw(G, pos=pos, with_labels=True,
        node_color="red", node_size=3000,
        font_color="white", font_size=20, font_family="Times New Roman",
        edge_color="black", width=5)
        
# add node labels to graph
nx.draw_networkx_labels(G, pos=pos_node_attributes, font_color="black", font_size=20, font_family="Times New Roman")
plt.margins(0.2)
plt.show()
```

```python
G = nx.Graph()
G.graph["Name"] = "My Graph"

G.add_nodes_from([
    ("A", {"Age":19, "Gender":"F"}),
    ("B", {"Age":18, "Gender":"M"}),
    ("C", {"Age":22, "Gender":"M"}),
    ("D", {"Age":21, "Gender":"M"}),
    ("E", {"Age":20, "Gender":"F"})
])

G.add_edges_from([
    ("A","C",{"weight":1}),
    ("B","C",{"weight":0.5}),
    ("B","D",{"weight":0.6}),
    ("C","D",{"weight":0.8}),
    ("D","E",{"weight":1})
])

pos = {
    "A" : (1, 5),
    "B": (4.5, 6.6),
    "C": (3.6, 1.4),
    "D": (5.8, 3.5),
    "E": (7.9, 3.6)
} # positions are defined as x and y coordinates

# define the positions of the node labels
# - handtyping the positions of the node labels
pos_node_attributes = {
    "A": (-0.2, 5),
    "B": (4.5, 7.5),
    "C": (2.4, 1.4),
    "D": (5.8, 2.5),
    "E": (9.1, 3.6)
} # these are the handtyped actual positions (x.y) coordinates of the node labels we want to add to the graph

# for n,d in G.nodes(data=True):
#     print(n)
#     print(d)
#     print("=========================")

node_labels = {n:(d["Age"],d["Gender"]) for n,d in G.nodes(data=True)}

edge_labels = {(u,v):d["weight"] for u,v,d in G.edges(data=True)}

# draw graph
nx.draw(G, pos=pos, with_labels=True,
        node_color="red", node_size=3000,
        font_color="white", font_size=20, font_family="Times New Roman",
        edge_color="black", width=5)
        
# add node labels to graph
nx.draw_networkx_labels(G, pos=pos_node_attributes, labels=node_labels,font_color="black", font_size=12, font_family="Times New Roman")
nx.draw_networkx_edge_labels(G, pos=pos, edge_labels=edge_labels, label_pos=0.5)

plt.margins(0.2)
plt.show()
```

```python
G = nx.Graph()
G.graph["Name"] = "My Graph"

G.add_nodes_from([
    ("A", {"Age":19, "Gender":"F"}),
    ("B", {"Age":18, "Gender":"M"}),
    ("C", {"Age":22, "Gender":"M"}),
    ("D", {"Age":21, "Gender":"M"}),
    ("E", {"Age":20, "Gender":"F"})
])

G.add_edges_from([
    ("A","C",{"weight":1}),
    ("B","C",{"weight":0.5}),
    ("B","D",{"weight":0.6}),
    ("C","D",{"weight":0.8}),
    ("D","E",{"weight":1})
])

pos = {
    "A" : (1, 5),
    "B": (4.5, 6.6),
    "C": (3.6, 1.4),
    "D": (5.8, 3.5),
    "E": (7.9, 3.6)
} # positions are defined as x and y coordinates

# define the positions of the node labels
# - handtyping the positions of the node labels
pos_node_attributes = {}

for node, (x,y) in pos.items():
    pos_node_attributes[node] = (x, y-0.9)


node_labels = {n:(d["Age"],d["Gender"]) for n,d in G.nodes(data=True)}

edge_labels = {(u,v):d["weight"] for u,v,d in G.edges(data=True)}

# draw graph
nx.draw(G, pos=pos, with_labels=True,
        node_color="red", node_size=3000,
        font_color="white", font_size=20, font_family="Times New Roman",
        edge_color="black", width=5)
        
# add node labels to graph
nx.draw_networkx_labels(G, pos=pos_node_attributes, labels=node_labels,font_color="black", font_size=12, font_family="Times New Roman")
nx.draw_networkx_edge_labels(G, pos=pos, edge_labels=edge_labels, label_pos=0.5)

plt.margins(0.2)
plt.show()
```

solve the maximum flow problem...

```python
G = nx.DiGraph() # define a directed graph
G.add_nodes_from(["S","A","B","C","T"])

# for the max flow problem...edge attribute ought to be named as capacity..alone
G.add_edges_from([
    ("S","A", {"capacity": 12}),
    ("S","B", {"capacity": 8}),
    ("S","C", {"capacity": 5}),
    ("A","B", {"capacity": 10}),
    ("A","C", {"capacity": 9}),
    ("B","T", {"capacity": 5}),
    ("C","B", {"capacity": 7}),
    ("C","T", {"capacity": 16})
])

from networkx.algorithms.flow import shortest_augmenting_path

flow_value, flow_dict = nx.maximum_flow(G, "S","T",flow_func=shortest_augmenting_path)

print(flow_value)*
print("====================================")
print(flow_dict)
```
