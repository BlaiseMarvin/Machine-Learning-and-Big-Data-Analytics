```python
import os
```

```python
rpath = r"C:\Users\blais\Documents\ML\data\cat_dog_images"
```

```python
file_list = []
```

```python
for file in os.listdir(rpath):
    if file.endswith('.jpg'):
        full_path = os.path.join(rpath,file)
        file_list.append(full_path)
```

```python
sorted(file_list)
```

Enter Pathlib: Object-Oriented file paths:
- pathlib treats paths as objects, not strings. this makes code readable, chainable and less error-prone

```python
import pathlib
```

```python
imgdir_path = pathlib.Path(rpath) # create the path object
```

```python
print(imgdir_path)
```

```python
print(type(imgdir_path))
```

```python
print(isinstance(imgdir_path,str))
```

Path Parts: No more string parsing

```python
p = pathlib.Path('folder/subfolder/file.txt')
```

```python
print(p.name)
```

```python
print(p.stem)
```

```python
print(p.suffix)
```

```python
print(p.parent)
```

```python
print(p.parts)
```

```python
print(p.parent.parent)
```

```python
print(p.parent/'other.txt')
```
