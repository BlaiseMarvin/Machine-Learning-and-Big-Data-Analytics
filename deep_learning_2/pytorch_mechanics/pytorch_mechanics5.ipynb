```python
import torch
import torch.nn as nn
import numpy as np
```

```python
def softmax(x):
    return np.exp(x)/np.sum(np.exp(x),axis=0)
```

```python
x = np.array([2.0,1.0,0.1])
outputs = softmax(x)
print('softmax numpy: ', outputs)
```

```python
x = torch.tensor([2.0,1.0,0.1])
torch_softmax=nn.Softmax(dim=0)
outputs = torch_softmax(x)
print('softmax pytorch: ', outputs)
```

Cross Entropy:

```python
def cross_entropy(actual,predicted):
    loss = -np.sum((actual*np.log(predicted) + (1-actual)*np.log(1-predicted)))
    return loss
```

all entropy under the hood is binary cross entropy - so for a multiclass classification problem - this translates to one-hot encoding the raw labels

```python
# if class 0: [1, 0, 0]
# if class 1: [0, 1, 1]
# if class 2: [0, 0, 1]
```

```python
Y = np.array([1, 0, 0])
```

```python
# y_pred has probabilities

Y_pred_good = np.array([0.7, 0.2, 0.1])
Y_pred_bad = np.array([0.1, 0.3, 0.6])
```

```python
l1 = cross_entropy(Y, Y_pred_good)
print(f'Loss1 numpy: {l1:.4f}')
```

```python
l2 = cross_entropy(Y, Y_pred_bad)
print(f'Loss2 numpy: {l2:.4f}')
```

Entropy in pytorch:

```python
loss = nn.CrossEntropyLoss()

Y = torch.tensor([2,0,1])

# nsamples x nclasses = 1x3
Y_pred_good = torch.tensor([[0.1, 1.0, 2.1],[2.0,1.0,0.1],[0.1,3.0,0.1]])
Y_pred_bad = torch.tensor([[2.1,1.0,0.1],[0.1,1.0,2.1],[0.1,3.0,0.1]])
```

```python
l1 = loss(Y_pred_good,Y)
l2 = loss(Y_pred_bad, Y)

print(f'L1 loss: {l1.item():.4f}')
print(f'L2 loss: {l2.item():.4f}')
```

```python
_, predictions1 = torch.max(Y_pred_good,1)
_, predictions2 = torch.max(Y_pred_bad, 1)
```

```python
print(predictions1)
print(predictions2)
```
