```python
import torch
import torchvision
from torch.utils.data import Dataset, DataLoader
import numpy as np
import math
```

```python
class WineDataset(Dataset):
    def __init__(self):
        # calling init methods from the inherited class - just in case
        super().__init__()
        # data loading
        xy = np.loadtxt('../../data/wine.csv', delimiter=',', dtype=np.float32, skiprows=1)
        self.x = torch.from_numpy(xy[:,1:]).clone()
        self.y = torch.from_numpy(xy[:,[0]]).clone() # n_samples
        self.n_samples = xy.shape[0]
    
    def __getitem__(self, index):
        return self.x[index],self.y[index]
    
    def __len__(self):
        return self.n_samples
```

```python
dataset = WineDataset()
```

```python
first_sample = dataset[0]
```

```python
dataloader = DataLoader(dataset=dataset,batch_size=4,shuffle=True)
```

```python
# training loop
num_epochs = 2
total_samples = len(dataset)
n_iterations = math.ceil(total_samples/4) # 4 is the batch size
print(total_samples, n_iterations)
```

```python
for epoch in range(num_epochs):
    for i, (inputs,labels) in enumerate(dataloader):
        # forward pass, backward pass, and update weights
        if (i+1)%5==0:
            print(f'epoch {epoch+1}/{num_epochs}, step {i+1}/{n_iterations}, inputs {inputs.shape}')
```
