**scikit-learn Transformers:**
- tsfresh includes 3 scikit learn compatible transformers - which allow you to easily incorporate feature extraction and feature selection from time series into existing ML pipelines
- The scikit-learn pipeline allows you to assemble several pre-processing steps that will be executed in sequence and thus can be cross validated together while setting different parameters. TSFresh transformers allow to extract and filter the time series features during the pre-processing sequence. 

- First 2 estimators in tsfresh are the FeatureAugmenter, which extracts the features, and feature selector which performs the feature selection algorithm. It is preferable to combine extracting and filtering of the features in a single step. hence the relevantfeatureaugmenter estimator.

Combining tsfresh's RelevantFeatureAugmenter and a RandomForestClassifier into a single pipeline. This pipeline can then fit both the transformer and classifier in one step

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from tsfresh.examples import load_robot_execution_failures
from tsfresh.transformers import RelevantFeatureAugmenter
import pandas as pd
```

```python
# download the dataset
from tsfresh.examples.robot_execution_failures import download_robot_execution_failures
download_robot_execution_failures()
```

```python
pipeline = Pipeline([
    ('augmenter', RelevantFeatureAugmenter(column_id = 'id', column_sort='time')),
    ('classifier', RandomForestClassifier())
])
```

```python
df_ts, y = load_robot_execution_failures()
```

```python
X = pd.DataFrame(index=y.index)
```

```python
pipeline.set_params(augmenter__timeseries_container=df_ts)
```

```python
pipeline.fit(X,y)
```
